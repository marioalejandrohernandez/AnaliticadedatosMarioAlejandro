```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .excel sin indice

import pandas as pd
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name="2023", header=3)
df
```

```python
from google.colab import files

df.to_csv("df.csv")
files.download("df.csv")
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
data1 =df.fillna(method="bfill")
valores_nulos=data1.isnull().sum()


valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
y = df
limite_superior = y.mean() + 3*y.std()
limite_inferior = y.mean() - 3*y.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

**Procedimiento para la desviacion estandar**

```python
Limite_Superior= data1.mean() + 3*data1.std()
Limite_Inferior= data1.mean() - 3*data1.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python

data2= data1[(data1<=Limite_Superior)&(data1>=Limite_Inferior)]
data3= data2.drop(['Otros'], axis=1)
valores_nulos=data3.isnull().sum()
valores_nulos

```

```python
df_1 = df[(y <= limite_superior) & (y >=limite_inferior)]
df_1
```

```python
from google.colab import files

df_1.to_csv("df1.csv")
files.download("df.csv")
```

```python
df_2 = df_1.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"], axis=1)
df_2
```

```python
from google.colab import files

df_2.to_csv("df2.csv")
files.download("df.csv")
```

```python
df_2.info()
```

```python
df_c = df_2.copy()
df_c = df_c.fillna(round(df_2.mean(),1))
df_c
```

```python
from google.colab import files

df_c.to_csv("dfc.csv")
files.download("df.csv")
```

```python
df_c.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df_c)
plt.suptitle("Outliers con desv")
plt.show()
```

```python
y = df_c

per25 = y.quantile(0.25) #Q1
per75 = y.quantile(0.75) #Q3

iqr = per75 - per75

lim_sup_iqr = per75+ 1.5*iqr
lim_inf_iqr = per25- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print("\nLímite inferior permitido" , lim_inf_iqr)
```

```python
#datos limpios de df
df_qr = df_c[(y <=lim_sup_iqr)&(y >=lim_inf_iqr)]
df_qr
```

```python
from google.colab import files

df_qr.to_csv("dfQR.csv")
files.download("df.csv")
```

```python
#remplazamos valores atípicos
df_qr2 = df_qr.copy()
df_qr2 = df_qr2.fillna(round(df_qr.mean(),1))
df_qr2
```

```python
from google.colab import files

df_qr2.to_csv("dfQR2.csv")
files.download("df.csv")
```

```python

columnas = [ "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"]
indice= df[columnas]
datos_limpios = pd.concat([indice, df_qr2],axis=1)
datos_limpios
```

```python
datos_limpios.to_csv("Datos limpios 2023.csv")
```
